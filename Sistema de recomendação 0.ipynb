### Recomendador baseado em popularidade

#### Criao de um subconjunto com produtos avaliados mais de 50 vezes. Este ser o nosso limite

```python
# Dataframe com produtos com mais de 50 avaliaes
new_df = dfOrders.groupby("product_id").filter(lambda x: x['review_score'].count() >= 50)

# # Criao de um novo DataFrame com classificao mdia e nmero de classificaes por produto
ratings_df = pd.DataFrame(new_df.groupby('product_id').review_score.mean())

# # Adicionando coluna com nmero de avaliaes por produto
ratings_df['rating_counts'] = new_df.groupby('product_id').review_score.count()

# # Visualizao dos 5 produtos mais bem avaliados (nmero de avaliaes)
ratings_df.sort_values(by = 'rating_counts', ascending = False).head(5)
```

Em seguida, definiremos uma funo para calcular uma mdia ponderada que pega o nmero total de votos e o limite (50) e o multiplica pela mdia global e pela mdia do produto. A frmula : (v / (v + m) * R) + (m / (m + v) * C)

- (v) votos do produto;
- (m) votos mnimos GERAL;

- (R) avaliao mdia do produto;
- (C) avaliao GERAL de produtos

```python
# Mdia Global das review_score
C = ratings_df['review_score'].mean()

# Limite mnimo para ser elegvel ao ranking
m = ratings_df.rating_counts.min()

# Funo que calcula a mdia ponderada de cada item
def weighted_rating (x, m = m, C = C):
  v = x['rating_counts']
  R = x['review_score']
  # Calcula mdia ponderada
  return (v/(v+m) * R) + (m/(v+m) * C)

# Adiciona a 'pontuao' calculada com weighted_rating() ao dataframe
ratings_df['score'] = ratings_df.apply(weighted_rating, axis=1)

# Resultado final com os 15 produtos mais populares
ratings_df.sort_values(by='score', ascending=False).head(15)
```
